Examine the NA value and outliers

```python
(raw_dataset.isnull().sum()/len(raw_dataset))
```

```python
df = raw_dataset.copy()
df['rear_camera_mp']=df['rear_camera_mp'].fillna(df['rear_camera_mp'].mean())
df['front_camera_mp']=df['front_camera_mp'].fillna(df['front_camera_mp'].mean())
df['internal_memory']=df['internal_memory'].fillna(df['internal_memory'].mean())
df['ram']=df['ram'].fillna(df['ram'].mean())
df['battery']=df['battery'].fillna(df['battery'].mean())
df['weight']=df['weight'].fillna(df['weight'].mean())
```

```python
df.isnull().sum()/len(df)
```

Remove the predicting target 'normalized_used_price'

```python
vis_df = df.copy()
```

```python
vis_df = vis_df.drop(['normalized_used_price'], axis = 1)
vis_df = vis_df.drop(['device_brand'], axis = 1)
vis_df = vis_df.drop(['os','4g','5g'], axis = 1)
```
